```python
bins = [0, 100000,300000, float('inf')]
labels = ['low','Med', 'High']
df ['income band'] = pd.cut(df['Estimated Income'],bins = bins, labels = labels, right = False)
```

```python
df['income band'].value_counts()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns

for i in labels:
    plt.figure()
    plt.title(f"Estimated Income - '{i}':")
    sns.histplot(data =df[df['income band'] == i ], x = 'Estimated Income', kde = True)
   
    
```

```python
#Displaying the value counts for the categorical data in the table 

Category_Columns = df[["BRId","GenderId","IAId","Amount of Credit Cards","Nationality", "Occupation", "Fee Structure", "Loyalty Classification",
                     "Properties Owned", "Risk Weighting","income band"]].columns
for col in Category_Columns:
    print(f"Values for {col}:")
    display(df[col].value_counts())
```

```python

for i, charts in enumerate(Category_Columns):
    plt.figure(i)
    sns.countplot(data = df, x = charts, hue = 'Nationality', palette = 'YlGnBu')
    plt.title(charts)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show

    

```

```python
#Numeric analysis
numerical_cols = df[['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits',
                     'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']].columns

plt.figure(figsize=(15, 10))

for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)  # i + 1 to start from subplot 1
    sns.histplot(df[col], kde=True)
    plt.title(col)
    plt.tight_layout()  # Prevent overlap

plt.show()  # Show all subplots together


```
